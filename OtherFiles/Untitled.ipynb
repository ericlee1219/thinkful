```python
import pandas as pd
import numpy as np
import os
print os.getcwd()
```

```python
all_ages = pd.read_csv('all-ages.csv')
all_ages.head(5)
recent_grads = pd.read_csv("recent-grads.csv")
print(all_ages.head(5))
print(recent_grads.head(5))
```

```python
from collections import defaultdict
```

```python
major_categories = all_ages["Major_category"].unique()
print(major_categories)

all_ages_major_categories = defaultdict(dict)
recent_grads_major_categories = defaultdict(dict)

for major in major_categories:
    is_major = all_ages["Major_category"] == major
    major_data = all_ages[is_major]
    major_data_total = major_data["Total"].sum()
    all_ages_major_categories[major] = major_data_total

for major in major_categories:
    is_major = recent_grads["Major_category"] == major
    major_data = recent_grads[is_major]
    major_data_total = major_data["Total"].sum()
    recent_grads_major_categories[major] = major_data_total
    
print(all_ages_major_categories)
    

```

```python
#Use the Low_wage_jobs and Total columns to calculate the proportion of recent college graduates 
#that worked low wage jobs. Store the resulting float as low_wage_percent.

Low_wage_jobs_total = recent_grads["Low_wage_jobs"].sum()
Total_grads = recent_grads["Total"].sum()
low_wage_percent = float(Low_wage_jobs_total / Total_grads) * 100
print(low_wage_percent)
```

```python
#Instructions
#We want to know the number of majors where recent grads fare better than the overall population.
#For each major:
#increment recent_grads_lower_emp_count if Unemployment_rate is lower for recent_grads
#increment all_ages_lower_emp_count if Unemployment_rate is lower for all_ages
#do nothing if Unemployment_rate is the same for both
recent_grads_lower_emp_count = 0
recent_grads_lower_emp_count_majors = []
all_ages_lower_emp_count = 0
all_ages_lower_emp_count_majors = []
majors = recent_grads["Major"].unique()
for major in majors:
    is_major_all_ages = all_ages["Major"] == major
    is_major_recent_grads = recent_grads["Major"] == major
    major_all_ages = all_ages[is_major_all_ages]
    major_recent_grads = recent_grads[is_major_recent_grads]
    if major_all_ages["Unemployment_rate"].sum() < major_recent_grads["Unemployment_rate"].sum():
        all_ages_lower_emp_count += 1
        all_ages_lower_emp_count_majors.append(major)
    elif major_all_ages["Unemployment_rate"].sum() > major_recent_grads["Unemployment_rate"].sum():
        recent_grads_lower_emp_count += 1
        recent_grads_lower_emp_count_majors.append(major)
    else:
        pass

print(recent_grads_lower_emp_count)
print(recent_grads_lower_emp_count_majors)
print(all_ages_lower_emp_count)
print(all_ages_lower_emp_count_majors)

```

```python
from pandas import Series

number_of_women = recent_grads["Women"].values

series_women_in_major = Series(number_of_women , index=majors)
print(series_women_in_major[['ANIMAL SCIENCES', 'FOOD SCIENCE']])
```
